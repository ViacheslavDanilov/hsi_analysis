## Test on raw data

```python
threshold = 0.50
img_dir = 'data/raw_converted/absorbance/30_08_2019_test_01_liver/10_00_39_T6=110'
save_dir = 'data/demo_prediction_raw'
```

```python
img_paths = get_file_list(
    src_dirs=img_dir,
    ext_list='.png',
)

for img_path in img_paths:
    save_path = os.path.join(save_dir, Path(img_path).name)
    detection = inference_detector(model, img_path)
    show_result_pyplot(
        model=model,
        img=img_path,
        result=detection,
        score_thr=threshold,
        title='Prediction: Source data',
        wait_time=0,
        palette=None,
        out_file=save_path,
    )
```
